```
# knock21

import pandas as pd

uselog = pd.read_csv('/content/drive/My Drive/sample_100knocks.zip (Unzipped Files)/ƒTƒ“ƒvƒ‹ƒR[ƒh_20200325/3Í/use_log.csv')
display(uselog.head(15))
display(uselog.info())

customer = pd.read_csv('/content/drive/My Drive/sample_100knocks.zip (Unzipped Files)/ƒTƒ“ƒvƒ‹ƒR[ƒh_20200325/3Í/customer_master.csv')
display(customer.head(15))
display(customer.info())

class_master = pd.read_csv('/content/drive/My Drive/sample_100knocks.zip (Unzipped Files)/ƒTƒ“ƒvƒ‹ƒR[ƒh_20200325/3Í/class_master.csv')
display(class_master.head(15))
display(class_master.info())

campaign_master = pd.read_csv('/content/drive/My Drive/sample_100knocks.zip (Unzipped Files)/ƒTƒ“ƒvƒ‹ƒR[ƒh_20200325/3Í/campaign_master.csv')
display(campaign_master.head(15))
display(campaign_master.info())
```

```
# knock22

# 顧客データの整形
customer_join = pd.merge(customer, class_master, on='class', how='left')
customer_join = pd.merge(customer_join, campaign_master, on='campaign_id', how='left')
display(customer_join.head())
display(customer_join.isnull().sum())
```

```
# 今年度の入会ユーザを集計してみる

# datetime化
customer_join['start_date'] = pd.to_datetime(customer_join['start_date'])

# loc関数に列のリストを渡すと、行の参照が可能
# https://note.nkmk.me/python-pandas-at-iat-loc-iloc/

# datetime化したテキストと比較演算子を使って比較
# 合致したデータをloc関数で該当するデータを抽出する -> なくてもいいんじゃないか
customer_start = customer_join.loc[customer_join['start_date'] > pd.to_datetime('20180401')]
customer_start2 = customer_join[customer_join['start_date'] > pd.to_datetime('20180401')]
display(customer_start)
display(customer_start2)

print(len(customer_start))

```

```
# knock24

# 現状の顧客データの分析

customer_join['end_date'] = pd.to_datetime(customer_join["end_date"])
customer_newer = customer_join.loc[(customer_join['end_date'] >= pd.to_datetime('20190331'))|(customer_join['end_date'].isna())]
display(customer_newer)
print(len(customer_newer))


customer_newer2 = customer_join[customer_join['end_date'] >= pd.to_datetime('20190331')]
customer_newer3 = customer_join[customer_join['end_date'].isna()]
display(customer_newer2)
display(customer_newer3)

print(customer_newer.groupby('class_name').count()['customer_id'])
print(customer_newer.groupby('campaign_name').count()['customer_id'])
print(customer_newer.groupby('gender').count()['customer_id'])
```

```
# knock 25

uselog['usedate'] = pd.to_datetime(uselog['usedate'])
uselog['年月'] = uselog['usedate'].dt.strftime('%Y%m')
uselog_month = uselog.groupby(['年月','customer_id'], as_index=False).count()
display(uselog_month)

uselog_month.rename(columns={'log_out':'count'}, inplace=True)
del uselog_month['usedate']
display(uselog_month)
```

```
uselog_customer = uselog_month.groupby('customer_id').agg(['mean', 'median', 'max', 'min'])
uselog_customer = uselog_customer.reset_index(drop=False)
uselog_customer
```

```
# knock26

uselog['weekday'] = uselog['usedate'].dt.weekday
uselog

uselog_weekday = uselog.groupby(['customer_id', '年月', 'weekday'], as_index=False).count()[['customer_id', '年月', 'weekday', 'log_id']]
uselog_weekday

uselog_weekday.rename(columns={'log_id':'count'}, inplace=True)
uselog_weekday
```

# matplotlib


## line

```
import pandas as pd
import numpy as np

np.random.seed(0)
val = np.random.randint(-5, 5, size=500).reshape(100, 5)
# cumsum -> 要素を足し合わせたものを、配列として出力する
df = pd.DataFrame(val, columns=list('abcde')).cumsum()
df
```

```
# 要素が複数あるときは、自動的に凡例がつく
df.plot(kind='line')
```

```
df['a'].plot(kind='line')
```

```
df['b'].plot(title='cplumn b line',
             legend=True)
```

```
df['c'].plot(grid=True,
             xlim=(0,20), # xのlimit
             ylim=(-30,30)) # yのlimit
```

```
# xticksにnp.arrange（初項, 終点, 公差）を渡すと目盛間隔の調整。grid=Trueとセット
df['b'].plot(xticks=np.arange(0,100,5),
             grid=True)
```

```
# 画像サイズ
df.plot(figsize=(3,2)) # インチ換算
```

```
# subplotsを渡すと並べて表示. sub
df[['a', 'b']].plot(subplots=True)
```

## bar

```
# ilocで1行目、全列のデータを参照
display(df.iloc[1,:])
df.iloc[1,:].plot(kind='bar')
```

```
# ilocで0から9行目、全列を参照
display(df.iloc[0:10,:])

# 複数の行がある場合はそれぞれの列をまとめて出力
df.iloc[0:10,:].plot(kind='bar')
```

```
df.iloc[0:5, :].plot(kind='bar',
                     stacked=True)
```

```
df.iloc[0:5, :].plot(kind='barh',
                     stacked=True)
```
